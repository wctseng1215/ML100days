```python
import pandas as pd
import numpy as np

boston_data=np.genfromtxt("boston.csv", delimiter=",",names=True)
boston_data
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?

#箱型圖
df = pd.DataFrame(boston_data)
print(df)
df.boxplot()
print()
print('TAX 中位數在300~400之間 ')
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?



#散佈圖
df = pd.DataFrame(boston_data)
print(df)
df.plot.scatter(x='NOX', y='DIS')
```
