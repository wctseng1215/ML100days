1.   畫出 cos 圖檔，並儲存
	plt.savefig("filename.png",dpi=300,format="png")

```python
# 載入需要的...

import matplotlib.pyplot as plt
import numpy as np

# 準備數據 ... 假設我要畫一個con波 從0~720度

x = np.arange(0,720)
y = np.cos(x * np.pi / 180.0)

# 開始畫圖
# 設定要畫的的x,y數據list....
# 如果要存成圖形檔:
# 把 pyplot.show() 換成下面這行:
    
    
plt.plot(x,y) 

plt.xlabel("x-axis") 
plt.ylabel("y-axis") 
plt.title("cos plot") 


plt.savefig("filename.png",dpi=300,format="png")
```
